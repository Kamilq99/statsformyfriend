```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Dane z tabeli
dane = {
    "Nazwa firmy": [
        "IZO-ERG SA w Gliwicach",
        "Huta Ferrum SA w Katowicach",
        "Zakłady Lentex SA w Lublińcu",
        "Wielkopolskie Przedsiębiorstwo Przemysłu Ziemniaczanego SA w Luboniu k/Poznania",
        "Zakłady Chemiczne Luboo SA",
        "Izolacja - Jarocin SA",
        "Ortal SA w Łodzi"
    ],
    "Atrakcyjność rynku": [320, 320, 370, 290, 360, 200, 350],
    "Pozycja konkurencyjna": [330, 280, 310, 340, 215, 220, 290]
}
```

```python
# Konwersja danych na DataFrame
df = pd.DataFrame(dane)

# Wygenerowanie wykresu
plt.figure(figsize=(10, 6))
plt.plot(df["Pozycja konkurencyjna"], df["Atrakcyjność rynku"], marker='o', linestyle=' ', c='blue', ms=10)

# Adnotacje dla wartości
for i, row in df.iterrows():
    # Przesunięcie o 10 pikseli w pionie
    y_pos = row['Atrakcyjność rynku'] + 10
    x_pos = row['Pozycja konkurencyjna']

    plt.annotate(f"({row['Pozycja konkurencyjna']},{row['Atrakcyjność rynku']})",
                  xy=(x_pos, y_pos),
                  ha='center', va='bottom', fontsize=7, color='black')

plt.xlabel("Pozycja konkurencyjna")
plt.ylabel("Atrakcyjność rynku")
plt.ylim(bottom=0, top=450)
plt.xlim(left=500, right=0)
plt.title("Atrakcyjność rynku w zależności od pozycji konkurencyjnej")
plt.grid(True)
plt.show()
```
